### Pandas + Matplotlib

```python
import pandas as pd

csv = '/content/sample/anime_stock_returns.csv'

df = pd.read_csv(csv,index_col='Date',parse_dates=["Date"])
df.head()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='NanumBarunGothic')
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)
ax.plot(df.index,df['TOEI ANIMATION'],label='TOEI ANIMATION')
ax = fig.add_subplot(111)
ax.plot(df.index,df['IG Port'],label='IG Port') # label  plot

ax.legend(loc='lower right') # legend   

ax.set_title("  2 ")
ax.set_xlabel("")
ax.set_ylabel("")

plt.xticks(rotation=45) # x  45  plot
plt.show()
```

```python
import pandas as pd

csv = '/content/sample/4816.csv'

df = pd.read_csv(csv,index_col=0,parse_dates=['Date'])

df.head()
```

Close()   Volume()  plot

```python
fig = plt.figure(figsize=(10,4))
ax1 = fig.add_subplot(111)
ax1.plot(df.index,df['Close'],color='b',label='')

ax2 = ax1.twinx() #    x   
ax2.bar(df.index,df['Volume'],color='g',label='',width=2)
ax1.legend()
ax1.set_xlabel("/")
ax1.set_yticks([i * 2000 for i in range(5)])

ax2.legend(loc=6)
ax2.set_yticks([i*5000 for i in range(5)])
ax2.set_ylabel("")

plt.show()
```

```python
csv = '/content/sample/anime_master.csv'

df = pd.read_csv(csv,index_col='anime_id')

df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(df['members'],df['rating'],alpha=0.5)
plt.figure(figsize=(15,15))
plt.show()
```

```python
# members  80  data 
df[df['members']>=800000]

df.loc[df['members']>=800000, ['name','members']]
```

```python
types = df['type'].unique()
types
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)

for t in types:
  x = df.loc[df['type']==t,'members'] # df['type'] t x members 
  y = df.loc[df['type']==t,'rating'] # df['type'] t y rating 
  ax.scatter(x,y,alpha=0.5,label=t)
  plt.legend(loc='lower right',fontsize=12)
plt.show()

```
